```python
x=df.drop('Price',axis=1)
y=df['Price']
```

```python
# train test 
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
```

```python
x_train=ss.fit_transform(xtrain)
```

```python
x_test=ss.transform(xtest)
```

```python
import pickle
pickle.dump(ss,open('scaling.pkl','wb'))
```

```python

x_train=pd.DataFrame(x_train,columns=xtrain.columns)
x_train
```

```python

x_test=pd.DataFrame(x_test,columns=xtest.columns)
x_test
```

```python
# Model Training
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,ytrain)
```

```python
# print the coefficients and the intercept
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
ypred=regression.predict(x_test)
```

```python
# plot a scatter for the prediction b/w ytest and ypred
plt.scatter(ytest,ypred)
plt.show()
```

```python
# Normality
resid=ytest-ypred
resid.skew()
```

```python
resid.plot(kind='kde')
```

```python
# uniform distribution b/w ypred and ypredict
plt.scatter(ypred,resid)
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
print('mae',mean_absolute_error(ytest,ypred))
print('mse',mean_squared_error(ytest,ypred))
print('rmse',np.sqrt(mean_squared_error(ytest,ypred)))
```

```python
print(r2_score(ytest,ypred))
```
